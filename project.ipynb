```python
# cleaning the data part 1;
# adding the coordinaes having the locations to each DataFrame 
#import Dependencies
import pandas as pd 
from geopy.geocoders import Nominatim
import geopandas as gpd
import re
def getcoordinates(my_df):
    ## define a funnction to realize if the input location is in string format or coordinate format
    def specialCharacters(s):
        special_characters = "°′″"
        if any(c in special_characters for c in s):
            response=True
            return response
        else:
            response=False
            return response
    # define a function to realize the input location as strings like (53°20′35″N 114°18′11″W﻿ / ﻿53.34306°N 114.30306°W')
    def twocoordinates(s):
        special_characters = "/"
        if any(c in special_characters for c in s):
            response=True
            return response
        else:
            response=False
            return response
    # turning dms coordinate to decimal coordinate

    def dmstodecimal(c):
        deg, minutes, seconds, direction =  re.split('[°′″]', c)
        dd=(float(deg) + float(minutes)/60 + float(seconds)/(60*60)) * (-1 if direction in ['W', 'S'] else 1)
        return dd

    # define a function to return lat and lon from the wiered coordinates 
    def stringtolatlong(string):
    # string like (53°20′35″N 114°18′11″W﻿ / ﻿53.34306°N 114.30306°W')
        if twocoordinates(string)==True:
            lat=float(string.split("W﻿ / ﻿")[1].split("°N")[0])
            lon=float(string.split("W﻿ / ﻿")[1].split("°N")[1].strip("°W"))*-1
            return(lat, lon)
        else:
    #  or ([[Grande Prairie 55°20′20″N 119°13′07″W]])
            try:
                l=string.split(" ")
                n=len(l)
                dmslat=l[n-2]
                lat=dmstodecimal(dmslat)
                dmslon=l[n-1]
                lon=dmstodecimal(dmslon)
                return (lat, lon)
    # or even (Forestburg 52°28′08″N112°08′02″W)
            except ValueError:
                m=string.split("″N")
                string=m[0]+"″N "+m[1]
                l=string.split(" ")
                n=len(l)
                dmslat=l[n-2]
                lat=dmstodecimal(dmslat)
                dmslon=l[n-1]
                lon=dmstodecimal(dmslon)
                return(lat, lon)
    # converting the location (like 'University of Calgary') to coordinates
    def findCoordinates(myaddress):
        try:
            geolocator = Nominatim(user_agent="my_user_agent")
            location = geolocator.geocode(myaddress)
            lon= location.longitude
            lat= location.latitude
    #         print(lat)
    #         print(lon)
            return(lat ,lon)
        except (TypeError, AttributeError,KeyError):
            pass

    # converting the location of the datatframe into coordinates
    latitudelist=[]
    longitudelist=[]
    for j in my_df['Location']:

        if specialCharacters(j)== True:
            print(j)
            print(stringtolatlong(j)[0])
            print(stringtolatlong(j)[1])
            latitudelist.append(stringtolatlong(j)[0])
            longitudelist.append(stringtolatlong(j)[1])
            next

        elif specialCharacters(j)== False:
            print(j)
            try:
                latitudelist.append(findCoordinates(j+' Canada')[0])
                print(findCoordinates(j+' Canada')[0])
                longitudelist.append(findCoordinates(j+' Canada')[1])
                print(findCoordinates(j+' Canada')[1])
            except(TypeError, AttributeError):
                latitudelist.append('NAN')
                longitudelist.append('NAN')
                pass
            next

    # Using DataFrame.insert() to add a column 
    my_df.insert(2, "lat", latitudelist, True) 
    my_df.insert(3, "lon", longitudelist, True)
```
